```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

## 2. Important ports

```python
# List of ports along with their coordinates
ports = {
    'Visakhapatnam': (17.7, 83.3),
    'Chennai': (13.1, 80.3),
    'Paradeep': (20.3, 86.6),
    'Kakinada': (16.9, 82.2),
    'Mumbai': (18.9, 72.8),
    'Kochi': (9.9, 76.3),
    'Mangalore': (12.9, 74.9),
    'Jawaharlal Nehru Port': (18.9, 72.8),
    'Surat Port':(21.1, 72.8),
    'Goa Port': (15.5, 73.9),
    'Kandla': (23.0, 70.0),
    'Tuticorin': (8.8, 78.2),
    'Port Blair': (11.6, 92.7),
}
```

### 2.1 Ports in India map


```python
# Set up the map with Basemap
plt.figure(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=6, urcrnrlat=23, llcrnrlon=70, urcrnrlon=93, resolution='i')


m.drawcoastlines()
m.drawcountries()
m.drawparallels(range(6, 23, 2), labels=[1, 0, 0, 0])
m.drawmeridians(range(70, 93, 2), labels=[0, 0, 0, 1])

# Plot ports and their bounding boxes
for port, (lat, lon) in ports.items():
    x, y = m(lon, lat)  # Convert lat/lon to map projection
    m.plot(x, y, marker='o', color='red', markersize=8)  # Plot port location

    # Annotate port name near the point
    plt.text(x + 50000, y, port, fontsize=8, color='black', weight='bold')  # Adjust offset for better visibility

# Display the map
plt.title('Ports along the Indian Coastline with 0.5° x 0.5° Bounding Boxes')
plt.show()
```

### 2.2 Port with grids


```python
# Set up the map with Basemap
plt.figure(figsize=(10, 8))
m = Basemap(projection='merc', llcrnrlat=6, urcrnrlat=24, llcrnrlon=66, urcrnrlon=93, resolution='i')

m.drawcoastlines()
m.drawcountries()
m.drawparallels(range(6, 24, 2), labels=[1, 0, 0, 0])
m.drawmeridians(range(66, 93, 2), labels=[0, 0, 0, 1])

# Plot ports and their bounding boxes
for port, (lat, lon) in ports.items():
    x, y = m(lon, lat)  # Convert lat/lon to map projection
    m.plot(x, y, marker='o', color='red', markersize=8)  # Plot port location

    # Annotate port name near the point
    plt.text(x + 50000, y, port, fontsize=8, color='black', weight='bold')  # Adjust offset for better visibility

    # Convert corners of the 0.5° x 0.5° box to map projection
    lon_min, lon_max = lon - 0.25, lon + 0.25
    lat_min, lat_max = lat - 0.25, lat + 0.25
    x_min, y_min = m(lon_min, lat_min)
    x_max, y_max = m(lon_max, lat_max)

    # Draw 0.5° x 0.5° box
    m.plot([x_min, x_max], [y_min, y_min], color='blue', linewidth=1)  # Bottom edge
    m.plot([x_min, x_max], [y_max, y_max], color='blue', linewidth=1)  # Top edge
    m.plot([x_min, x_min], [y_min, y_max], color='blue', linewidth=1)  # Left edge
    m.plot([x_max, x_max], [y_min, y_max], color='blue', linewidth=1)  # Right edge

# Display the map
plt.title('Ports along the Indian Coastline with 0.5° x 0.5° Bounding Boxes')
plt.show()
```
